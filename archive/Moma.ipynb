## Data cleaning

```python
specific_columns = df[['Artist', 'BeginDate', 'EndDate', 'Nationality',
                            'Gender',  'Date', 'Height (cm)', 'Width (cm)',]]
specific_columns
```

```python
specific_columns.columns = ['artist', 'birth', 'death', 'nationality', 'gender', 'painting_date', 'height', 'width']

specific_columns
```

```python
specific_columns_copy = specific_columns.copy()

specific_columns_copy.to_csv('artworks_specific_columns.csv', index=False)

print("specific columns copy 'artworks_specific_columns.csv'.")
```

```python
columns_to_clean = ['nationality', 'gender', 'birth', 'death']
# remove parentheses
for column in columns_to_clean:
    specific_columns_copy[column] = specific_columns_copy[column].str.replace(r'[()]', '', regex=True)

specific_columns_copy
```

```python
country_counts = specific_columns_copy['nationality'].value_counts()
country_counts
```

```python
import matplotlib.pyplot as plt

# plot the top 10 countries
country_counts.head(100).plot(kind='bar', figsize=(30, 6), title='Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Records')
plt.show()
```

```python
country_counts = specific_columns_copy['nationality'].value_counts()

one_word_countries = country_counts[country_counts.index.str.match(r'^\w+$')]

print(one_word_countries)

one_word_countries.to_csv('Artworks_Cleaned.csv')
print("One-word country saved 'Artworks_Cleaned.csv'.")
```

```python
import matplotlib.pyplot as plt

# plot all countries
one_word_countries.head(1000).plot(kind='bar', figsize=(30, 6), title='Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Records')
plt.show()
```

```python
#gender count

gender_counts = specific_columns_copy['gender'].value_counts()

print(gender_counts.head(20))

gender_counts.to_csv('Gender_Counts.csv')
print("Gender counts saved 'Gender_Counts.csv'.")
```

```python
# male and female 
def clean_gender(value):
    value = str(value).lower().strip()  
    words = value.split() 
    contains_male = 'male' in words  
    contains_female = 'female' in words 

    if contains_male and contains_female:
        return 'unknown'  
    elif contains_male:
        return 'male'
    elif contains_female:
        return 'female'
    else:
        return 'unknown' 

# applying to the column
specific_columns_copy['gender'] = specific_columns_copy['gender'].apply(clean_gender)

gender_counts_cleaned = specific_columns_copy['gender'].value_counts()
gender_counts_cleaned
```

```python
# converting all the dataset to lower case
specific_columns_copy = specific_columns_copy.applymap(lambda x: str(x).lower() if isinstance(x, str) else x)

print(specific_columns_copy.head())
```

```python
specific_columns_copy.info()
```

```python
artworks_data = specific_columns_copy.copy()

# columns to fill with 'unknown'
columns_unknown = ['artist', 'nationality', 'gender']

specific_columns_copy[columns_unknown] = specific_columns_copy[columns_unknown].applymap(
    lambda x: 'unknown' if pd.isna(x) or str(x).strip() == '' else x
)

# columns to fill with 0
columns_zero = ['birth', 'death', 'painting_date', 'height', 'width']

specific_columns_copy[columns_zero] = specific_columns_copy[columns_zero].applymap(
    lambda x: 0 if pd.isna(x) or str(x).strip() == '' else x
)

artworks_data[columns_unknown] = artworks_data[columns_unknown].astype(str)

# converting numeric columns to numeric data type
artworks_data[columns_zero] = artworks_data[columns_zero].apply(pd.to_numeric, errors='coerce')

artworks_data.info()
```

```python
artworks_data.to_csv('artworks_data.csv', index=False)

print("Updated dataset saved as 'artworks_data.csv'.")
```

## 2.	What are the top 10 most represented nationalities among the artists?

```python
# nationality is NaN and filter out 'unknown', 'nan', and 'null'
filtered_nationalities = artworks_data[
    artworks_data['nationality'].notna() & 
    ~artworks_data['nationality'].str.lower().isin(['unknown', 'nan', 'null', ''])
]

# top 10 nationalities excluding invalid values
top_nationalities_filtered = filtered_nationalities['nationality'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_nationalities_filtered.plot(
    kind='bar',
    color=['#FFB3BA', '#BAE1FF', '#FFDFBA', '#B3FFD9', '#D9BAFF', 
           '#FFE1BA', '#FFBABA', '#BAFFD9', '#FFD9BA', '#BABAFF']
)
plt.title('Top 10 Most Represented Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Number of Artists')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3.	What is the average lifespan of the artists?

```python
artworks_data['lifespan'] = artworks_data['death'] - artworks_data['birth']

valid_lifespans = artworks_data.loc[(artworks_data['death'] > 0) & (artworks_data['lifespan'] > 0), 'lifespan']

average_lifespan = int(valid_lifespans.mean())

print(f"Average lifespan is: {average_lifespan} years")
```

## 4.	How has the representation of male vs. female artists changed over time?

```python
gender_representation_over_time = artworks_data.groupby(['painting_date', 'gender']).size().unstack(fill_value=0)

gender_representation_filtered = gender_representation_over_time[['male', 'female']]

plt.figure(figsize=(10, 5))
gender_representation_filtered.plot(
    kind='line',
    title='Representation of Male vs. Female Artists Over Time'
)
plt.xlabel('Year')
plt.ylabel('Number of Artists')
plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

The graph highlights the consistent dominance of male artists in MoMA’s collection, with representation peaking in the mid-20th century. This reflects historical biases in the art world, where male artists were more recognized and supported. Female artists are significantly underrepresented, with a modest increase in their representation during the 20th century, though their numbers remain far below those of male artists. Post-2000, acquisitions for both male and female artists decline sharply, likely due to changes in MoMA’s acquisition strategy, financial limitations, and a shift toward non-traditional or digital art forms.

## 5.	What is the average age of artists at the time their artworks were created?

```python
cleaned_artworks_data = artworks_data.dropna().copy() 

# using loc due: not to modify the sliced dataframe
cleaned_artworks_data.loc[:, 'age_at_creation'] = cleaned_artworks_data['painting_date'] - cleaned_artworks_data['birth']

valid_ages = cleaned_artworks_data.loc[
    (cleaned_artworks_data['age_at_creation'] > 15) & 
    (cleaned_artworks_data['age_at_creation'] <= 100), 
    'age_at_creation'
]

average_age_at_creation = int(valid_ages.mean())

print(f"Average age of artists at the time of artwork creation: {average_age_at_creation} years")
```

Average age of artists at the time of artwork creation is 44 years.

## 7.	How diverse is MoMa in terms of represented nationalities?

```python
import pandas as pd

# top 30 countries
data = {
    'nationality': [
        'American', 'French', 'German', 'British', 'Spanish', 'Italian', 'Japanese', 'Russian',
        'Swiss', 'Dutch', 'Belgian', 'Mexican', 'Austrian', 'Argentine', 'Brazilian', 'Canadian',
        'Czech', 'Colombian', 'Chilean', 'Polish', 'Venezuelan', 'Danish', 'Ivorian', 'Chinese',
        'Israeli', 'Swedish', 'Australian', 'Indian', 'Finnish', 'Cuban',
    ],
    'count': [
        73547, 22882, 9637, 5693, 3157, 2786, 2465, 2159, 2124, 1615, 1478, 1478, 925, 922, 915, 789,
        778, 773, 608, 560, 529, 527, 482, 322, 316, 300, 253, 230, 220, 219,
    ]
}
df = pd.DataFrame(data)

# non european/usa countries in top 30
excluded_countries = [
    'Japanese', 'Mexican', 'Argentine', 'Brazilian', 'Colombian', 'Chilean', 'Venezuelan',
    'Ivorian', 'Chinese', 'Israeli', 'Australian', 'Indian', 'Cuban'
]

excluded_total = df[df['nationality'].isin(excluded_countries)]['count'].sum()
included_total = df[~df['nationality'].isin(excluded_countries)]['count'].sum()

percentage_included = (included_total / excluded_total) * 100

print(f"Non European countries total: {excluded_total}")
print(f"European/Usa countries total: {included_total}")
print(f"Percentage of European/Usa relative to non european countries: {percentage_included:.2f}%")
```

```python
import matplotlib.pyplot as plt

categories = ['Non-European/USA Countries', 'European/USA Countries']
totals = [excluded_total, included_total]
colors = ['#BAE1FF', '#FFB3BA']  # Pastel colors

plt.figure(figsize=(6, 6))
plt.pie(
    totals,
    labels=categories,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    textprops={'fontsize': 12}
)

plt.title('Percentage of European/USA vs Non-European/USA Countries in Artwork Count', fontsize=14)
plt.tight_layout()

plt.show()
```

MoMA’s collection is heavily skewed towards American and European nationalities, with 1358.04% more artworks from these regions compared to non-European nationalities within the top 30. This indicates that the museum’s representation of diverse global cultures is significantly limited.

## 8.	Are there differences in the average lifespan of artists across nationalities? 


```python
# removing null or empty nationalities
cleaned_artworks = artworks_data.dropna(subset=['birth', 'death', 'nationality'])
cleaned_artworks = cleaned_artworks[cleaned_artworks['nationality'].str.strip() != '']

# lifespan for each artist
cleaned_artworks['lifespan'] = cleaned_artworks['death'] - cleaned_artworks['birth']

# invalid lifespans (less than 15 or greater than 100 years)
valid_lifespans = cleaned_artworks[
    (cleaned_artworks['lifespan'] >= 15) & (cleaned_artworks['lifespan'] <= 100)
]

# group by nationality and calculate the average lifespan
average_lifespan_by_nationality = valid_lifespans.groupby('nationality')['lifespan'].mean().round().astype(int)

print("Average Lifespan of Artists by Nationality:")
print(average_lifespan_by_nationality)
```

## 9. Are there noticeable differences between male and female artists in terms of age at creation, number of works, or representation by nationality?


```python
cleaned_artworks = artworks_data.dropna(subset=['birth', 'painting_date', 'gender', 'nationality'])
cleaned_artworks = cleaned_artworks[
    (cleaned_artworks['gender'].str.strip() != '') &
    (cleaned_artworks['nationality'].str.strip() != '')
]

cleaned_artworks['age_at_creation'] = cleaned_artworks['painting_date'] - cleaned_artworks['birth']

valid_ages = cleaned_artworks[cleaned_artworks['age_at_creation'] >= 15]

average_age_by_gender = valid_ages.groupby('gender')['age_at_creation'].mean().round(2)

gender_representation_by_nationality = valid_ages.groupby(['nationality', 'gender']).size().unstack(fill_value=0)

gender_representation_percentage = gender_representation_by_nationality.div(
    gender_representation_by_nationality.sum(axis=1), axis=0
) * 100

gender_representation_percentage = gender_representation_percentage.round(2)

print("Average Age at Creation by Gender:")
print(average_age_by_gender)

print("\nGender Representation Percentage by Nationality:")
print(gender_representation_percentage)
```

```python
top_10_nationalities = gender_representation_percentage.head(20)

top_10_nationalities.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    color=['#BAE1FF', '#FFB3BA', '#FFD9BA'],
    edgecolor='black'
)
plt.title('Gender Representation Percentage by Nationality (Top 20)')
plt.xlabel('Nationality')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

## 10. What is the average, smallest, and largest size (area) of the artworks?


```python
import pandas as pd

cleaned_artworks = artworks_data.dropna(subset=['height', 'width'])
cleaned_artworks = cleaned_artworks[
    (cleaned_artworks['height'] > 0) & (cleaned_artworks['width'] > 0)
]

# height and width from cm to meters
cleaned_artworks['height_m'] = cleaned_artworks['height'] / 100
cleaned_artworks['width_m'] = cleaned_artworks['width'] / 100

# area of each artwork in square meters
cleaned_artworks['area_m2'] = cleaned_artworks['height_m'] * cleaned_artworks['width_m']

# exclude artworks with an area of zero
cleaned_artworks = cleaned_artworks[cleaned_artworks['area_m2'] > 0]

average_area_m2 = cleaned_artworks['area_m2'].mean()
smallest_area_m2 = cleaned_artworks['area_m2'].min()
largest_area_m2 = cleaned_artworks['area_m2'].max()

print(f"Average Area: {average_area_m2:.2f} m²")
print(f"Smallest Area: {smallest_area_m2:.2f} m²")
print(f"Largest Area: {largest_area_m2:.2f} m²")
```

## 11. Is there a correlation between the size of artworks and the artist's gender or year?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cleaned_artworks = artworks_data.dropna(subset=['height', 'width', 'gender', 'painting_date', 'nationality'])
cleaned_artworks = cleaned_artworks[
    (cleaned_artworks['height'] > 0) & (cleaned_artworks['width'] > 0)
]

# height and width to meters and calculate area
cleaned_artworks['height_m'] = cleaned_artworks['height'] / 100
cleaned_artworks['width_m'] = cleaned_artworks['width'] / 100
cleaned_artworks['area_m2'] = cleaned_artworks['height_m'] * cleaned_artworks['width_m']

# correlation with gender
gender_area = cleaned_artworks.groupby('gender')['area_m2'].mean()

# correlation with year (painting_date)
year_area = cleaned_artworks.groupby('painting_date')['area_m2'].mean()

plt.figure(figsize=(8, 6))
gender_area.plot(kind='bar', color=['#BAE1FF', '#FFB3BA'], edgecolor='black')
plt.title('Average Artwork Size by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Artwork Size (m²)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=year_area, marker="o", color="#FFA07A")
plt.title('Average Artwork Size by Year')
plt.xlabel('Year')
plt.ylabel('Average Artwork Size (m²)')
plt.tight_layout()
plt.show()

```
